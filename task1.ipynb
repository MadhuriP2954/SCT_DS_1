```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_1292964.csv',header=2)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
data=df.set_index('Country Name').iloc[:, 4:].T
data=data.apply(pd.to_numeric,errors='coerce').dropna(how='all')
data
```

```python
total=data.sum(axis=1)
plt.figure(figsize=(15,6))
total.plot(kind='bar')
plt.xlabel=('Year')
plt.ylabel=('Total population')
plt.title("Total population by year")
plt.show()
```
